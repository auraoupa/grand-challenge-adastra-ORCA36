```python
import numpy as np
import xarray as xr

import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.util as cutil

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

import cmocean
import dask
```

```python
ds=xr.open_dataset('/gpfsstore/rech/cli/rote001/ETOPO1_Ice_g_gmt4.nc4')
ds
```

```python
ds.z.max()
```

```python
topo=ds.z
lon=ds.x
lat=ds.y
```

```python
projection=ccrs.Orthographic(central_latitude=30.0, central_longitude=-30.0)
fig, ax = plt.subplots(subplot_kw=dict(projection=projection), figsize=(10,9))
ax.set_global()

#-- add coastal outlines
ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth=0.5)
#ax.background_img(name='BM', resolution='low')

gl = ax.gridlines(draw_labels=True, linewidth=0.5, color='k', zorder=3)
gl.xlabel_style = {'size':10}
gl.ylabel_style = {'size':10}
gl.top_labels   = True
gl.right_labels = True

cnf1  = ax.pcolormesh(lon, lat, topo,
                          cmap=cmocean.cm.gray,
                          vmin=0,
                          vmax=8000,
                          shading='flat',
                          transform=ccrs.PlateCarree())
```
